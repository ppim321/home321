```python
import folium

# 여러 위치 정보를 담은 리스트 (1~5개의 좌표)
locations = [
    {"name": "불국사", "coords": [35.7892734269, 129.3318475148]},
    {"name": "첨성대", "coords": [35.834457, 129.219747]},
    {"name": "석굴암", "coords": [35.799152, 129.331462]},
    {"name": "경주 대릉원", "coords": [35.842823, 129.211559]},
    {"name": "동궁과 월지", "coords": [35.834874, 129.227047]},
]

# 지도 생성 (첫 번째 위치 기준)
map_folium = folium.Map(location=locations[0]["coords"], zoom_start=14)

# 각 위치에 Marker 추가
for location in locations:
    folium.Marker(
        location=location["coords"],
        popup=location["name"],  # 마커 클릭 시 이름 표시
        icon=folium.Icon(color="red", icon="info-sign")
    ).add_to(map_folium)

# 지도 출력
map_folium
```

```python
import folium
from geopy.geocoders import Nominatim

# 지명을 입력받아 지도에 표시하는 함수
def create_map_with_locations():
    # 사용자로부터 1~5개의 지명을 입력받기
    print("여정에 추가할 장소를 최대 5개 입력하세요 (예: 불국사, 첨성대). 입력을 마치면 Enter를 눌러주세요.")
    locations = []
    for i in range(5):
        location = input(f"지명 {i+1} (입력을 종료하려면 Enter): ").strip()
        if not location:
            break
        locations.append(location)

    # 지명이 없는 경우
    if not locations:
        print("지명을 입력하지 않았습니다.")
        return None

    # Geopy 지오코더 초기화
    geolocator = Nominatim(user_agent="geo_folium")

    # 지도 초기 위치는 첫 번째 지명
    first_location = geolocator.geocode(locations[0])
    if not first_location:
        print(f"'{locations[0]}'의 위치를 찾을 수 없습니다.")
        return None

    # 지도 생성
    map_folium = folium.Map(location=[first_location.latitude, first_location.longitude], zoom_start=14)

    # 입력된 지명으로 마커 추가
    for location_name in locations:
        location = geolocator.geocode(location_name)
        if location:
            folium.Marker(
                location=[location.latitude, location.longitude],
                popup=location_name,  # 마커 클릭 시 지명 표시
                icon=folium.Icon(color="red", icon="info-sign")
            ).add_to(map_folium)
        else:
            print(f"'{location_name}'의 위치를 찾을 수 없습니다.")

    # 지도 반환
    return map_folium

# 함수 실행
map_with_locations = create_map_with_locations()
if map_with_locations:
    map_with_locations.save("map_with_locations.html")
    print("지도를 'map_with_locations.html'로 저장했습니다.")
```

```python
# 저장한 파일 불러오기

import webbrowser
import os

# 저장된 HTML 파일을 열어주는 함수
def show_map(file_name="map_with_locations.html"):
    # 파일 경로 확인
    if os.path.exists(file_name):
        # 브라우저로 파일 열기
        webbrowser.open(f"file://{os.path.abspath(file_name)}")
        print(f"'{file_name}' 파일을 브라우저에서 열었습니다.")
    else:
        print(f"'{file_name}' 파일을 찾을 수 없습니다. 먼저 지도를 생성하세요.")

# 실행
show_map()
```

```python
import folium
from geopy.geocoders import Nominatim

# 지명을 입력받아 지도에 표시하고 경로를 그리는 함수
def create_map_with_locations_and_route():
    # 사용자로부터 1~5개의 지명을 입력받기
    print("최대 5개의 지명을 입력하세요 (예: 불국사, 첨성대). 입력을 마치면 Enter를 눌러주세요.")
    locations = []
    for i in range(5):
        location = input(f"지명 {i+1} (입력을 종료하려면 Enter): ").strip()
        if not location:
            break
        locations.append(location)

    # 지명이 없는 경우
    if not locations:
        print("지명을 입력하지 않았습니다.")
        return None

    # Geopy 지오코더 초기화
    geolocator = Nominatim(user_agent="geo_folium")

    # 위치 정보 리스트
    coordinates = []

    # 입력된 지명으로 위도와 경도를 가져옴
    for location_name in locations:
        location = geolocator.geocode(location_name)
        if location:
            coordinates.append([location.latitude, location.longitude])  # 위도, 경도 저장
        else:
            print(f"'{location_name}'의 위치를 찾을 수 없습니다.")

    # 위치 정보가 없으면 종료
    if not coordinates:
        print("유효한 위치 정보가 없습니다.")
        return None

    # 지도 생성 (첫 번째 위치 기준)
    map_folium = folium.Map(location=coordinates[0], zoom_start=14)

    # 마커 추가 및 경로 표시
    for idx, coord in enumerate(coordinates):
        folium.Marker(
            location=coord,
            popup=locations[idx],  # 마커 클릭 시 지명 표시
            icon=folium.Icon(color="red", icon="info-sign")
        ).add_to(map_folium)

    # 경로(폴리라인) 추가
    if len(coordinates) > 1:
        folium.PolyLine(
            locations=coordinates,
            color="blue",
            weight=5,
            opacity=0.7
        ).add_to(map_folium)

    # 지도 반환
    return map_folium

# 함수 실행
map_with_locations_and_route = create_map_with_locations_and_route()
if map_with_locations_and_route:
    map_with_locations_and_route.save("map_with_locations_and_route.html")
    print("지도를 'map_with_locations_and_route.html'로 저장했습니다.")
```
